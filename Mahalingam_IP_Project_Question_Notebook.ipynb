### **Question 13:** The company wants to provide a promotional offer in the advertisement of the restaurants. The condition to get the offer is that the restaurants must have a rating count of more than 50 and the average rating should be greater than 4. Find the restaurants fulfilling the criteria to get the promotional offer. [3 marks]

```python
# replacing the not given entries for the rating column to not a number
data['rating'] = data['rating'].replace('Not given', np.nan)

# Converting the column to numeric data type
data['rating'] = pd.to_numeric(data['rating'])

# Checking converted data of Rating column
data.info()
```

```python
# Grouping by restaurant name to compute mean rating
company_with_rating = data.groupby('restaurant_name').agg({'rating': ['count', 'mean']})
company_with_rating
# Renaming columns
company_with_rating.columns = ['rating_count', 'average_rating']

# Filtering restaurants based on criteria
promotional_restaurants = company_with_rating[(company_with_rating['rating_count'] > 50) & (company_with_rating['average_rating'] > 4)]

print("Restaurants fulfilling the criteria for the promotional offer:")
print(promotional_restaurants)
```

**Observations:**


*   The restaurants Blue Ribbon Fried Chicken, Blue Ribbon Sushi, Shake Shack, and The Meatball shop receive a rating above four on a avaerage.
*   The restaurant Shake Shack received the most number of ratings amongst the four.





### **Question 14:** The company charges the restaurant 25% on the orders having cost greater than 20 dollars and 15% on the orders having cost greater than 5 dollars. Find the net revenue generated by the company across all orders. [3 marks]

```python
# Computing the revenue for orders > 20 dollars
revenue_greater_than_20 = data[data['cost_of_the_order'] > 20]['cost_of_the_order'].sum() * 0.25

# Calculating the revenue for orders > 5 dollars and <= 20 dollars
revenue_greater_than_5 = data[(data['cost_of_the_order'] > 5) & (data['cost_of_the_order'] <= 20)]['cost_of_the_order'].sum() * 0.15

# Calculate total revenue
total_revenue = revenue_greater_than_20 + revenue_greater_than_5

print("Net revenue generated by the company: $", total_revenue, )
```

**Observations:**
*   Overall, the company made a profit of 6K USD, helping customers with delivery via the mobie app.
